```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

df_2004_use=df_2004.drop(['전국'])

sns.set(rc={'figure.figsize':(10,8)})
plt.hist(df_2004_use['선거구'],bins=5)
plt.xlabel("number of constituency")
plt.show()
```

```python
correlation_matrix=df_2004_use.corr().round(3)
sns.heatmap(data=correlation_matrix,annot=True, cmap='bwr')
```

```python
sns.set_style('darkgrid')
fig,axs=plt.subplots(ncols=3)
sns.regplot(x='인구밀도(명/㎢)',y='선거구 비율',data=df_2004_use,ax=axs[0])
sns.regplot(x='1인당 민간소비',y='선거구 비율',data=df_2004_use,ax=axs[1])
sns.regplot(x='인구',y='선거구 비율',data=df_2004_use,ax=axs[2])
```
