```
import random
x=random.uniform(-20,20)
print("first x value is",x)

def GD(x):
  for i in range(1000):
    y=x**2+1
    y_prime=2*x

    if(y<1.01):
      print("The Optimal Solution of X is ", x, "and y is at X is ", y, "and the number of calculation to find the solution is",i)
      break
    else:
      print("(",x,",",y,")")
      if(y_prime>0):
        x-=0.1
      else:
        x+=0.1


def NM(x):
  for i in range(1000):
    y=x**2+1
    y_prime=2*x

    if(y<1.01):
      print("The Optimal Solution of X is ", x, "and y is at X is ", y, "and the number of calculation to find the solution is",i)
      break
    else:
      print("(",x,",",y,")")
      x=x-y/y_prime

print("1. use Gradient Descent")
GD(x)
print()
print("2. use Newton method")
NM(x)
```
